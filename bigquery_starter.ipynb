# Big Query
Google Cloud has an uptodate nyc 311 data stored [here](https://bigquery.cloud.google.com/table/bigquery-public-data:new_york.311_service_requests)

This notebook runs a simpe query on the data.

You need a file called auth_key.json that has your google cloud service credentials

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
query = """
SELECT * FROM 
`bigquery-public-data.new_york.311_service_requests` 
LIMIT 1000

"""
```

```python
df = pd.read_gbq(query, project_id='np-hacks',verbose=True, private_key='auth_key.json', dialect='standard')
```
